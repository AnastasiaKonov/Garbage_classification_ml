```python
import torch
from torch.utils.data import DataLoader
from torch.utils.data import Dataset
from torch import nn
from torchvision import transforms
from torchvision.models import ResNet50_Weights, resnet50
from matplotlib import pyplot as plt
from PIL import Image
import os
import numpy as np
from tqdm.notebook import tqdm
```

```python
# declare constants

IMG_SIZE = 640
CLASS_NAMES = {'concrete':[1, 0, 0, 0], 'soil':[0, 1, 0, 0], 'wood':[0, 0, 1, 0], 'brick':[0, 0, 0, 1]}
DATA_PATH = '/content/drive/MyDrive/dataset'
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# define useful functions

def show_image(img):
  """
  img - torch.tensor, PIL.Image object or a path to an image file
  """
  if isinstance(img, torch.Tensor):
    plt.imshow(img.view(-1, IMG_SIZE, IMG_SIZE).permute(1, 2, 0).detach().cpu().numpy())
  elif isinstance(img, str):
    img = Image.open(img)
    plt.imshow(img)
  else:
    plt.imshow(img)
  plt.show()

def plot_history(history):
  plt.figure(figsize = (9, 6))
  epochs = list(range(len(history['train'])))
  plt.plot(epochs, history['train'], label = 'train')
  plt.plot(epochs, history['test'], label = 'test')
  plt.title('Средний лосс на эпоху')
  plt.show()
```

```python
model = resnet50(ResNet50_Weights.IMAGENET1K_V2) # инициализируем модель с предобученными весами

classifier = nn.Sequential(
    nn.Linear(2048, 1024),
    nn.ReLU(inplace=True),
    nn.Linear(1024, 512),
    nn.ReLU(inplace=True),
    nn.Linear(512, 256),
    nn.ReLU(inplace=True),
    nn.Linear(256, 4),
    nn.Softmax()
)
model.fc = classifier
```

```python
# define important stuff

optimizer = torch.optim.Adam(model.parameters())
'''
в нашем случае MSE будет считаться так:
пусть p = (p1, p2, p3, p4) - предсказанные нами вероятности каждого из 4 классов,
а Y = (y1, y2, y3, y4) - истинные значения, при чем среди них есть только одна единица, а все остальные значения - нули

поскольку эта информация берется из датасета, где мы изначально знаем, какой мусор на данной картинке,
вероятность этого класса будет 100% (т.е. единица), а остальные равны нулю, т.к. вероятности должны складываться в 1
MSE = ((y1-p1)^2 + (y2-p2)^2 + (y3-p3)^2 + (y3-p3)^2)/4

'''
criterion = nn.MSELoss() # средний квадрат ошибки

# аугментации
transf = transforms.Compose(
    # изначально делаем изображение чуть больше желаемного, чтобы потом после обрезки оно стало нужного размера
    [#transforms.Resize(
        #(int(IMG_SIZE * 1.15), int(IMG_SIZE * 1.15)), Image.BICUBIC),
     transforms.RandomCrop((IMG_SIZE, IMG_SIZE))
     transforms.RandomHorizontalFlip(),
     #transforms.RandomVerticalFlip(),
     transforms.ToTensor()]
)
```

```python
#dataLoader
class ImageDataset(Dataset):
  def __init__(self, root, transform, mode) -> None:
    """
    root - директория с датасетом.
    в ней должны быть папки train и test, в которых в свою очередь лежат тренировочный и тестовый наборы данных.
    сами наборы данных - это 4 папки с названиями 'concrete', 'soil', 'wood' и 'brick' для каждого класса соответственно

    transform - аугментации, объект класса torchvision.transforms.
    их я вроде уже сделал (лежат в переменной 'transf'), нужно только передать сюда параметром
    применяются только на этапе обучения, при тестировании и использовании - нет.

    mode - строка 'train' либо 'eval' - режим, в котором работает датасет.
    т.е. тренировочный или тестовый. от этого зависит применение аугментаций и директория, из которой берутся кртинки
    """
    self.mode = mode
    self.transforms = transforms.ToTensor()
    if self.mode == 'train':
      self.transforms = transform
    self.files = []
    self.labels = []

    image_set_name = 'train'
    if self.mode == 'test':
      image_set_name = 'test'
    if self.mode == 'valid':
      image_set_name = 'valid'

    dir = os.path.join(root, image_set_name)
    for class_dir in os.listdir(dir):
      f = os.path.join(dir, class_dir)
      for filename in os.listdir(f):
        full_path = os.path.join(f, filename)
        if os.path.isfile(full_path):
          self.files.append(full_path)
          self.labels.append(CLASS_NAMES[class_dir.lower()])

    assert len(self.labels) == len(self.files), "each file should have a label"

  def __getitem__(self, index):
    """
    возвращает кортеж, где 0-й элемент это само изображение в виде тензора,
    1-й элемент - вектор, с нулями на всех местах и еденицой на месте верного класса
    (можно посмотреть, как это выглядит в переменной 'CLASS_NAMES')
    """

    img = Image.open(self.files[index%len(self.files)])
    img = self.transforms(img)
    labels = self.labels[index%len(self.labels)]

    return (img, torch.tensor(labels, dtype=torch.float32))

  def __len__(self):
    """
    возвращает длину датасета
    """
    return len(self.files)
```

```python
def eval_model(model, data_path):
    val_loader = DataLoader(
        ImageDataset(data_path, transf, 'valid'),
        batch_size = 1,
        shuffle = True,
        num_workers = 2
    )
    loss_val = 0
    for img, label in val_loader:
        with torch.no_grad():
            img = img.to(DEVICE)
            label = label.to(DEVICE)

            preds = model(img)
            print(list(CLASS_NAMES.keys())[np.argmax(preds.detach().cpu())], list(CLASS_NAMES.keys())[np.argmax(label.detach().cpu())])
            loss = criterion(preds, label)

            loss_val += loss.item()
        torch.cuda.empty_cache()
    print(f'loss: {loss_val/len(val_loader)}')
```

```python
def train(model, data_path, epochs, batch_size):
  optimizer = torch.optim.Adam(model.parameters())
  template = 'epoch: [{:3}/{:3}]\tloss: [{:7.5f} (train)], [{:7.5f} (test)]'
  train_loader = DataLoader(
    ImageDataset(data_path, transf, 'train'),
    batch_size = batch_size,
    shuffle = True,
    num_workers = 2
  )

  test_loader = DataLoader(
      ImageDataset(data_path, transf, 'test'),
      batch_size = batch_size,
      shuffle = True,
      num_workers = 2
  )

  # история, чтобы потом посмотреть на красивые (надеюсь) графики
  history = {'train' : [], 'test' : []}

  #переводим модель на ГПУ
  model = model.to(DEVICE)
  for epoch in tqdm(range(epochs)):
    # переводим модель в режим обучения
    epoch_loss_train = 0
    model.train()
    for img, label in tqdm(train_loader):
      img = img.to(DEVICE)
      label = label.to(DEVICE)
      # основной этап обучения: предсказания -> лосс -> обратное распространение ошибки -> шаг оптимизатора
      preds = model(img)
      loss = criterion(preds, label)
      epoch_loss_train += loss.item()
      loss.backward()
      optimizer.step()
      optimizer.zero_grad()
      torch.cuda.empty_cache()
    history['train'].append(epoch_loss_train/len(train_loader))

    epoch_loss_test = 0
    # тестируем модель
    model.eval()
    for img, label in test_loader:
      with torch.no_grad():
        img = img.to(DEVICE)
        label = label.to(DEVICE)

        preds = model(img)
        loss = criterion(preds, label)

        epoch_loss_test += loss.item()
      torch.cuda.empty_cache()
    history['test'].append(epoch_loss_test/len(train_loader))
    print(template.format(epoch+1, epochs, epoch_loss_train/len(train_loader), epoch_loss_test/len(train_loader)))

  torch.save(model.state_dict(), f'latest_{epochs}epochs.pt')
  print(f'\n\nlatest version saved in latest_{epochs}epochs.pt')
  return history
```

```python
epochs = 2
batch_size = 1

history = train(model, DATA_PATH, epochs, batch_size)
```
